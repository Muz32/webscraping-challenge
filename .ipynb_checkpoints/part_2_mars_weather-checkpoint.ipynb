```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url= "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
table = soup.find('table', class_='table')
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list to store the extracted data
mars_data = []

# Find all table rows
table_rows = table.find_all('tr')

# Extract column headings from the first row (index 0)
table_headers = [th.text.strip() for th in table_rows[0].find_all("th")]

# Loop through the remaining rows (excluding the first row with column headings)
for row in table_rows[1:]:
    # Find all data cells in the current row
    data_cells = row.find_all("td")
    
    # Extract row data from the data cells
    row_data = [cell.text.strip() for cell in data_cells]
        
    # Create a dictionary for each row
    mars_dict = dict(zip(table_headers, row_data))
        
    # Add the dictionary to the list
    mars_data.append(mars_dict)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
mars_temperature_df=pd.DataFrame(mars_data)
```

```python
# Confirm DataFrame was created successfully
mars_temperature_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
print(mars_temperature_df.dtypes)
```

```python
# Change data types for data analysis
mars_temperature_df['id'] = mars_temperature_df['id'].astype('int64')
mars_temperature_df['terrestrial_date'] = pd.to_datetime(mars_temperature_df['terrestrial_date'])
mars_temperature_df['sol'] = mars_temperature_df['sol'].astype('int64')
mars_temperature_df['ls'] = mars_temperature_df['ls'].astype('int64')
mars_temperature_df['month'] = mars_temperature_df['month'].astype('int64')
mars_temperature_df['min_temp'] = mars_temperature_df['min_temp'].astype(float)
mars_temperature_df['pressure'] = mars_temperature_df['pressure'].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
print(mars_temperature_df.dtypes)
```

### Step 5: Analyse the Data

Analyse your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?

#Get the unique values in the 'month' column
unique_months = mars_temperature_df['month'].unique()

#Count the number of unique months
num_months = len(unique_months)

num_months
```

```python
# 2. How many Martian days' worth of data are there?
martian_days = len(mars_temperature_df['sol'].unique())
martian_days 
```

```python
# 3. What is the average low temperature by month?
average_min_temperature = mars_temperature_df.groupby ('month')['min_temp'].mean()
average_min_temperature 
```

```python
# Plot the average temperature by month
#Charting the data, giving it a title and labelling the axis. 
average_min_temperature_chart = average_min_temperature .plot(kind="bar", title="Average minimum temperature in Mars")
average_min_temperature_chart.set_xlabel("month")
average_min_temperature_chart.set_ylabel("Temperature in Celsius")

#showing the chart
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location
average_min_temperature_sorted = average_min_temperature.sort_values()
coldest_to_warmer_months_chart = average_min_temperature_sorted.plot(kind="bar", title="Average minimum temperature from coldest to warmest month in Mars")
coldest_to_warmer_months_chart.set_xlabel("month")
coldest_to_warmer_months_chart.set_ylabel("Temperature in Celsius")

#showing the chart
plt.show()
```

**Observation** From the above chart, it can be seen that on average, the third month has the coldest temperature while the eighth month had the warmest temperature on Mars. 

```python
# 4. Average pressure by Martian month
average_pressure = mars_temperature_df.groupby ('month')['pressure'].mean()
average_pressure 
```

```python
# Plot the average pressure by month
average_pressure_sorted = average_pressure.sort_values()
average_pressure_chart = average_pressure_sorted.plot(kind="bar", title="Average atmospheric pressure in Mars")
average_pressure_chart.set_xlabel("month")
average_pressure_chart.set_ylabel("Atmospheric pressure")

#showing the chart
plt.show()
```

**Observation** It can be noted from the chart that Atmospheric pressure, on average, was lowest in the sixth month and highest in the ninth month. 

```python
# 5. How many terrestrial (earth) days are there in a Martian year?

# Calculate the number of terrestrial days for each data point
mars_temperature_df['terrestrial_days'] = (mars_temperature_df['terrestrial_date'] - mars_temperature_df['terrestrial_date'].min()).dt.days

# Plotting the daily minimum temperature against the number of terrestrial days
plt.figure(figsize=(10, 6))
plt.plot(mars_temperature_df['terrestrial_days'], mars_temperature_df['min_temp'])
plt.title('Daily Minimum Temperature vs. Terrestrial Days')
plt.xlabel('Number of Terrestrial Days')
plt.ylabel('Minimum Temperature (Celsius)')
plt.show()

```

**Observation** The distance from peak to peak temperature is approximately 1728-1038, or 690 days. Therefore, according to the plot, a year in mars is approximatly 690 terrestrial days. 

```python
#To find the exact values for terrestrial days associated with the peak to peak temperature. 
# Define the intervals
interval_1 = (1000, 1250)
interval_2 = (1500, 2000)  

# Find the index of the maximum temperature in the first interval
index_max_temp_interval_1 = mars_temperature_df.loc[
    (mars_temperature_df['terrestrial_days'] >= interval_1[0]) & 
    (mars_temperature_df['terrestrial_days'] <= interval_1[1]), 
    'min_temp'
].idxmax()

# Find the index of the maximum temperature in the second interval
index_max_temp_interval_2 = mars_temperature_df.loc[
    (mars_temperature_df['terrestrial_days'] >= interval_2[0]) & 
    (mars_temperature_df['terrestrial_days'] <= interval_2[1]), 
    'min_temp'
].idxmax()

# Get the terrestrial day numbers for the max temperatures
terrestrial_day_max_temp_interval_1 = mars_temperature_df.loc[index_max_temp_interval_1, 'terrestrial_days']
terrestrial_day_max_temp_interval_2 = mars_temperature_df.loc[index_max_temp_interval_2, 'terrestrial_days']

# Print the results
print(f"The terrestrial day number corresponding to the max temperature between {interval_1[0]} to {interval_1[1]} is: {terrestrial_day_max_temp_interval_1}")
print(f"The terrestrial day number corresponding to the max temperature between {interval_2[0]} to {interval_2[1]} is: {terrestrial_day_max_temp_interval_2}")
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
